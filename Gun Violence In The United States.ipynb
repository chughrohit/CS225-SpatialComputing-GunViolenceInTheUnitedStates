## Compare Monthly Incidents by Year

### Filter Data by Year

```python
df_city_2015 = df_city[(df_city['date'] >= '2015-01-01') & (df_city['date'] <= '2015-12-31')]
df_city_2016 = df_city[(df_city['date'] >= '2016-01-01') & (df_city['date'] <= '2016-12-31')]
df_city_2017 = df_city[(df_city['date'] >= '2017-01-01') & (df_city['date'] <= '2017-12-31')]
df_city_2018 = df_city[(df_city['date'] >= '2018-01-01') & (df_city['date'] <= '2018-12-31')]
df_city_2019 = df_city[(df_city['date'] >= '2019-01-01') & (df_city['date'] <= '2019-12-31')]
df_city_2020 = df_city[(df_city['date'] >= '2020-01-01') & (df_city['date'] <= '2020-12-31')]
```

### Count Indicents for Filtered Data

```python
df_city_2015_monthly = df_city_2015['date'].groupby([df_city_2015['date'].dt.month.rename('month')]).agg({'count'})
df_city_2016_monthly = df_city_2016['date'].groupby([df_city_2016['date'].dt.month.rename('month')]).agg({'count'})
df_city_2017_monthly = df_city_2017['date'].groupby([df_city_2017['date'].dt.month.rename('month')]).agg({'count'})
df_city_2018_monthly = df_city_2018['date'].groupby([df_city_2018['date'].dt.month.rename('month')]).agg({'count'})
df_city_2019_monthly = df_city_2019['date'].groupby([df_city_2019['date'].dt.month.rename('month')]).agg({'count'})
df_city_2020_monthly = df_city_2020['date'].groupby([df_city_2020['date'].dt.month.rename('month')]).agg({'count'})
```

### Plot Monthly Data

```python
months = {
    1 : "Jan",
    2 : "Feb",
    3 : "Mar",
    4 : "Apr",
    5 : "May",
    6 : "Jun",
    7 : "Jul",
    8 : "Aug",
    9 : "Sep",
    10: "Oct",
    11: "Nov",
    12: "Dec"
}


label = 2015
for frame in [df_city_2015_monthly, df_city_2016_monthly, df_city_2017_monthly, df_city_2018_monthly, df_city_2019_monthly, df_city_2020_monthly]:
    plt.plot([months[idx] for idx in frame.index], \
             frame['count'], label=str(label))
    label += 1

plt.legend(loc="upper right")
plt.grid(b=True, which='major', color='grey', linestyle='dashdot')

plt.title("Gun Violence From Jan - Dec for Years 2015 - 2020")
plt.xticks(range(0, 12))
plt.xlabel('Number of Incidents', fontsize=20)
plt.ylabel('Months', fontsize=20)

plt.rcParams["figure.figsize"] = (20,10)
plt.show()
```

## Gun Laws by State ( 2014 - 2018 )

```python
'''
    Data Source : https://statefirearmlaws.org/national-data/2018
'''
state_laws_count_2014 = pd.read_json('data/state_laws/state_laws_count_2014.json', typ='series')
state_laws_count_2015 = pd.read_json('data/state_laws/state_laws_count_2015.json', typ='series')
state_laws_count_2016 = pd.read_json('data/state_laws/state_laws_count_2016.json', typ='series')
state_laws_count_2017 = pd.read_json('data/state_laws/state_laws_count_2017.json', typ='series')
state_laws_count_2018 = pd.read_json('data/state_laws/state_laws_count_2018.json', typ='series')
```

```python
y1, y2, y3, y4, y5 = [], [], [], [], []
x1 = []
for x, y in state_laws_count_2014.items():
    y1.append(y)
    y2.append(state_laws_count_2015[x])
    y3.append(state_laws_count_2016[x])
    y4.append(state_laws_count_2017[x])
    y5.append(state_laws_count_2018[x])
    x1.append(x)

counter1 = go.Bar(x=x1,y=y1,name='2014', marker=dict(color="#f27da6"))
counter2 = go.Bar(x=x1,y=y2,name='2015', marker=dict(color="#7f9bef"))
counter3 = go.Bar(x=x1,y=y3,name='2016', marker=dict(color="#94e8ba"))
counter4 = go.Bar(x=x1,y=y4,name='2017', marker=dict(color="#e8d54a"))
counter5 = go.Bar(x=x1,y=y5,name='2018', marker=dict(color="#B43757"))

layout = go.Layout(barmode='group', title="Rise of Gun Violence Laws (2014 - 2018)")
fig = go.Figure(data=[counter1, counter2, counter3, counter4, counter5], layout=layout)
iplot(fig, filename='State-Laws-Count')
```
